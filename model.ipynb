```python
# importing the necessary libraries
import numpy as np 
import pandas as pd
```

```python
# read the csv file 
# Tell the pandas to read the special characters as latin1 through encoding
data = pd.read_csv('laptop_price.csv',encoding='latin1')
```

```python
# checkign the sample data 
data.sample(10)
```

```python
# check the data types 
data.info()
```

```python
# checking is there any null values  ?
data.isnull().sum()
```

```python
# converting the Object type ram field into int type field and remove the GB part 
data['Ram'] = data['Ram'].astype('str')
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Ram']
```

```python
# converting the weight field into float field and remove the KG part 
data['Weight'] = data['Weight'].astype('str')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float64')
data['Weight']
```

```python
#find the correlation between laptopid , inches , ram , weight with Price_euros , and only take the numeric values 
data.corr(numeric_only=True)['Price_euros']
```

```python
#checking the categories- of company 
data['Company'].value_counts()
```

```python
# categorizing the lower value coloumns into a one column as other 
other_companies = {'Samsung', 'Mediacom', 'Razer', 'Microsoft', 'Vero', 
                   'Xiaomi', 'Chuwi', 'Fujitsu', 'Google', 'LG', 'Huawei'}
data['Company'] = data['Company'].apply(lambda x :'other' if x in other_companies else x)
data['Company'].value_counts()
```

```python
# checking the categories of product 
len(data['Product'].value_counts())
```

```python
# checking the categories of Type name 
data['TypeName'].value_counts()
```

```python
# checking the categories of Screen resolutions 
data['ScreenResolution'].value_counts()
```

```python
# dividing the entire columns into 2 parts as IPS and Touchscreen 
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x: 1 if 'Touchscreen' in x else 0 )
data['Ips'] = data['ScreenResolution'].apply(lambda x: 1 if 'IPS' in x else 0)
```

```python
# checking the categories of cpu 
data['Cpu'].value_counts()
```

```python
# only gettign the first 3 names of the cpu name column
data['Cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
data['Cpu_name'].value_counts()
```

```python
# dividie the cpu into categories 
data['Cpu_name'] = data['Cpu_name'].apply(
    lambda x: 'AMD' if 'AMD' in x 
    else 'intel core i7' if x == 'Intel Core i7' 
    else 'intel core i5' if x == 'Intel Core i5' 
    else 'intel core i3' if x == 'Intel Core i3' 
    else 'Other'
)

print(data['Cpu_name'].value_counts())
```

```python
# checking the cateogaries  of GPU 
data['Gpu'].value_counts()
```

```python
# Divide the GPU into categories 
data['GPU_Name'] = data['Gpu'].apply(lambda x:'Intel' if 'Intel' in x 
                                    else 'Nvidia' if 'Nvidia' in x
                                    else 'AMD' if 'AMD' in x 
                                    else 'other' )
data['GPU_Name'].value_counts()
```

```python
data.head()
```

```python
# checking the categories of Opsys
data['OpSys'].value_counts()
```

```python
# divide the operationg systems data 
data['Opsys_name'] = data['OpSys'].apply(lambda x:'Windows' if 'Windows' in x 
                                        else 'Linux' if 'Linux' in x
                                        else 'Mac' if 'Mac' or 'mac' in x 
                                        else 'other' )
```

```python
# checking the cateogary types 
data['Opsys_name'].value_counts()
```

```python
data.head()
```

```python
# drop the unwanted categaries
data = data.drop(columns = ['laptop_ID','Inches','ScreenResolution','Cpu','Gpu','OpSys','Product'])
```

```python
data.head()
```

```python
data.info()
```

```python
# change the entire table into one hot encoding 
data = pd.get_dummies(data,dtype=int)
```

```python
# check first 2 rows
data.head(2)
```

```python
data.shape
```

```python
# divide the features as  x and y(output) value  
x = data.drop('Price_euros',axis=1)
y = data['Price_euros']
```

```python
# divide the train and test values with test size into 0.25 
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(
    x, y, test_size=0.25
)
```

```python
# make a function to output the model score 
def acc_score(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test)
    print(str(model) + '-->' + str(acc))
```

```python
# deciding the best regression model 
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
acc_score(lr)

from sklearn.linear_model import Lasso 
ls = Lasso()
acc_score(ls)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
acc_score(dt)

from sklearn.ensemble import RandomForestRegressor 
rf = RandomForestRegressor()
acc_score(rf)
```

```python
# tuning hyper parameters 
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(x_train, y_train)

best_model = grid_fit.best_estimator_

best_model.score(x_test, y_test)
```

```python
# saving the model in a  pickle file 
import pickle 
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model,file)
```
