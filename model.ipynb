```python
# importing the libraries as 
import numpy as np 
import pandas as pd
```

```python
# read the csv file 
data = pd.read_csv('laptop_price.csv',encoding='latin1')
```

```python
# checkign the sample data 
data.sample(5)
```

```python
# check the data types 
data.info()
```

```python
# checking is there any null values  ?
data.isnull().sum()
```

```python
# converting the Object type ram field into int type field 
data['Ram'] = data['Ram'].astype('str')
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Ram']
```

```python
# converting the weight field into float field 
data['Weight'] = data['Weight'].astype('str')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float64')
data['Weight']
```
